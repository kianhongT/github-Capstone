# To create a dataframe for Toronto's neighbourhoods

```python
import requests # library to handle requests
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
toronto_url = requests.get(url)
toronto_url
```

```python
data = pd.read_html(toronto_url.text)
```

```python
# to check the number of dataframes and its type in the url

len(data), type(data)
```

```python
# to check if the first table in the url is the table I want

trt_data = data[0]
trt_data
```

```python
#  Ignore cells with a borough that is Not assigned

trt_data = trt_data[trt_data.Borough != 'Not assigned']
trt_data
```

```python
# to check if there is any Not assigned neighborhood

df = trt_data[trt_data['Neighborhood'].str.match('Not assigned')]
df
```

```python
# to combine neighbourhoods of the same Post Code

df3 = trt_data.groupby('Postal Code')['Neighborhood'].apply(','.join).reset_index()
df3
```

```python
new_data = trt_data.merge(df3)
new_data
```

```python
new_data.sort_values(by=['Postal Code'], inplace=True)
new_data
new_df = new_data.reset_index(drop=True)
new_df
```

```python
new_df.shape
print('Number of rows in the dataframe:',new_df.shape[0])
```

# Adding geographical coordinates of each postal code to the Dataframe

```python
# retrieving geographical coordinates csv file

url = 'http://cocl.us/Geospatial_data'
coordinates = pd.read_csv(url)
coordinates
```

```python
coordinates.sort_values(by=['Postal Code'], inplace=True)
coordinates
new_df2 = coordinates.reset_index(drop=True)
new_df2
```

```python
# merging the geographical coordinates dataframe with the original dataframe

new_df = new_df.merge(new_df2, on='Postal Code', how='left')
new_df
```

# Explore and cluster the neighborhoods in Toronto (only boroughs that contain the word Toronto)

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
# create new dataframe which contains Toronto in their Borough.

df4 = new_df[new_df['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[43.6532, -79.3832], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
# Using K-means for neighbourhoods clustering

k=4
toronto_clustering = df4.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_

# Assigning a cluster label to each neighbourhood

df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4
```

```python
# obtaining cluster centers

k_means_cluster_centers = kmeans.cluster_centers_
k_means_cluster_centers
```

```python
# create map
map_clusters = folium.Map(location=[43.6532, -79.3832],zoom_start=12)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
